````python
# Power Rule Derivative Calculator

"""
This script calculates the derivative of a term using the power rule.

To run this script on your local machine, follow these steps:

1. Save the Script to a File:
   - Copy this code and paste it into a text editor (e.g., Visual Studio Code, Cursor, etc).
   - Save the file with a `.py` extension, for example: `derivative_calculator.py`.

2. Open the Terminal:
   - Press `Cmd + Space` and type "Terminal" to open the terminal.

3. Navigate to the Directory:
   - Use the `cd` command to navigate to the directory where you saved the Python script.
   - For example, if you saved it in the `Documents` folder:
     ```
     cd ~/Documents
     ```

4. Run the Python Script:
   - Run the script using Python. On macOS, you can use either `python3` or `python`, depending on your installation:
     ```
     python3 derivative_calculator.py
     ```
   - This will execute the script, and you will be prompted to enter the coefficient and exponent.

5. Follow the Prompts:
   - Enter the values for `a` (the coefficient) and `n` (the exponent) when requested.
   - The program will then display the derivative of the given term.

Requirements:
- Python 3 must be installed on your machine.

GitHub: @liamkeyek

"""

# Function to calculate the derivative using the power rule.
def calculate_derivative(a, n):
    """
    Parameters:
    a (float): Coefficient of the term.
    n (int): Exponent of the term.
    
    Returns:
    tuple: New coefficient and new exponent after differentiation.
    """
    if n == 0:
        return 0, 0
    
    new_coefficient = a * n
    new_exponent = n - 1
    return new_coefficient, new_exponent

# Main function to prompt the user for input and display the derivative.
def main():
    """
    Main function to prompt the user for input and display the derivative.
    """
    print("Power Rule Derivative Calculator")
    
    try:
        a = float(input("Enter the coefficient (a): "))
        n = int(input("Enter the exponent (n): "))
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        return
    
    new_coefficient, new_exponent = calculate_derivative(a, n)
    
    if new_coefficient == 0:
        print("The derivative is 0.")
    elif new_exponent == 0:
        print(f"The derivative is {new_coefficient}")
    elif new_exponent == 1:
        print(f"The derivative is {new_coefficient}x")
    else:
        print(f"The derivative is {new_coefficient}x^{new_exponent}")

# Entry point of the script
if __name__ == "__main__":
    main()
````
